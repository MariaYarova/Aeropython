```python
def zhukovski (z, c):
    ksi = z + c**2/z
    
    return ksi
```

```python
def get_velocity_doublet(strength, xd, yd, X, Y):
    """
    Returns the velocity field generated by a doublet.
    """
    u = - strength/(2*numpy.pi)*((X-xd)**2-(Y-yd)**2)/\
    ((X-xd)**2+(Y-yd)**2)**2
    v = - strength/(2*math.pi)*2*(X-xd)*(Y-yd)/\
    ((X-xd)**2+(Y-yd)**2)**2
    
    return u, v
```

```python
def get_stream_function_doublet(strength, xd, yd, X, Y):
    """
    Returns the stream-function generated by a doublet.
    """
    psi = - strength/(2*math.pi)*(Y-yd)/((X-xd)**2+(Y-yd)**2)
    
    return psi
```

```python
R_c = 1.15
doublet_strength = 2*numpy.pi*(R_c**2)
u_inf = 1

psi_freestream = u_inf*Y
```

```python
psi_z = get_stream_function_doublet(doublet_strength, -0.15, 0, X, Y) + psi_freestream
```

```python
def get_velocity_vortex(strength, xv, yv, X, Y):
    
    u = + strength/(2*math.pi)*(Y-yv)/((X-xv)**2+(Y-yv)**2)
    v = - strength/(2*math.pi)*(X-xv)/((X-xv)**2+(Y-yv)**2)
    return u, v

def get_stream_function_vortex(strength, xv, yv, X, Y):
    psi = strength/(4*math.pi)*numpy.log((X-xv)**2+(Y-yv)**2)
    
    return psi
```

```python
y_arter_rotation = R_c*math.sin(AoA)

gamma = y_arter_rotation*4*math.pi

u_vortex, v_vortex = get_velocity_vortex(gamma, -0.15, 0, X, Y)
```
